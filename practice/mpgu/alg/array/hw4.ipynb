```python
# Изучить и написать, что такое set внутри(хэш-таблица)
# В Python set - это структура данных, которая представляет собой неупорядоченное множество уникальных элементов.
# В отличие от других структур, set не имеет дупликатов.

s = [1, 2, 3, 1]
print(f"List: {s}")
print(f"Set: {set(s)}")
```

```python
# Основной механизм работы set - хэш-таблица.
# Хэш-таблица — это структура данных, которая позволяет осуществлять операции вставки, 
# удаления и поиска элементов за время, близкое к O(1).

hash_set = set()
hash_set.add("blue") 

# Изначально у нас уже есть 5 bucket - страниц в хэш-таблице. Когда мы добавляем blue , 
# то создаем хэш-код, а затем вычисляется индекс:
# hash-code % hash-table 

hash_code = hash("blue")  
print(f"Hash-code элемента: {hash_code}")
index = hash_code % 5  # Определяем индекс в хэш-таблице с 5 бакетами
print(f"Индекс элемента: {index}")
# На место этого индекса встает элемент.
# Если bucket пуст, элемент просто размещается там. Если бакет уже занят, 
# то используется стратегия для разрешения коллизий (например, linear probing (LP)).
# Но LP может вызывать проблемы, поэтому есть счетчик его использования и при превышении
# этот метод заменяется другим.


# Когда количество элементов в set увеличивается, и они начинают занимать все больше бакетов,
# хэш-таблица автоматически расширяется, чтобы избежать слишком большого количества коллизий 
# и поддерживать высокую производительность.
```
